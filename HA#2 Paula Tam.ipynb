```python
#CSC620
#Paula Abigail Tam <921850992>
#HA2
#Due: September 7, 2022

from nltk.chat.util import Chat, reflections
from nltk.stem import PorterStemmer

pairs = (
    (
        r"[iI] don\'?t understand (.*)",
        (
            "I can send you some resources so you can better understand %1",
        ),
    ),
    (
        r"[bB]ecaus* I\'?m have (.*) in my class",
        (
            "Do you think you need a tutor?",
        ),
    ),
    (
        r"(.*) I need a tutor",
        (
            "I can send you some info about some tutors. These should help.",
        ),
    ),
    (
        r"[iI]\'?m worri about (.*)",
        (
            "Why are you worried about %1?",
            "Do you think you require help from a tutor?",
        ),
    ),
    (
        r"[iI]t wa* (.*)",
        (
            "Why do you say that they were %1?",
            "How are you doing grade wise?",
            
        ),
    ),
    (
        r"(.*) grade* are (.*)",
        (
            "Why do you think your grades are %2?",
            "Why would your grades be %2?",
        ),
    ),
    (
        r"[tT]hank you",
        (
            "It was no problem, did you want to talk about anything else?",
        ),
    ),
    (
        r"[yY]e*",
        (
            "Okay. What can I help you with?",
        ),
    ),
    (
        r"[nN]o*",
        (
            "Okay. Thank you for talking with me today! Type 'quit' to exit.",
        ),
    ),
    (
        r"[qQ]uit",
        (
            "Good-bye.",
            "Have a good day!",
        ),
    ),
    (
        r"(.*)",
        (
            "Please tell me more.",
            "Can you elaborate on that?",
        ),
    ),
)

stemmer = PorterStemmer()
dict = {}


def word_tokenizer_stem(line):
    tokens = line.split()
    stems = [stemmer.stem(t) for t in tokens] #for loop to iterate the list of tokens
    #stems is a list at this point
    joined_stems = " ".join(stems) #so it doesnt display as a list anymore
    for t in joined_stems.split(): #iterates the string of stems
        word_counter(t) #does the counting thing 
    return joined_stems
    

def word_counter(stem):
    if stem in dict:
        dict[stem] += 1 #increment count if in dict
    else:
        dict[stem] = 1 #adds new entry
        
        
def total_tokens():
    total = 0
    for value in dict.values():
        total += value
    return total

    
def mod_converse(self, quit="quit"):
    user_input = ""
    while user_input != quit:
        user_input = quit
        try:
            user_input = input(">")
            stemmed_input = word_tokenizer_stem(user_input)
        except EOFError:
            print(user_input)
        if stemmed_input:
            while stemmed_input[-1] in "!.":
                stemmed_input = stemmed_input[:-1]
            print("Stemmed Input: ", stemmed_input) #calls tokenizer+stem function and prints
            print(self.respond(stemmed_input))
    print("Unique Types: ", len(dict), "\nTotal Tokens: ", total_tokens())      
    
    
eliza_chatbot = Chat(pairs, reflections)
Chat.converse = mod_converse #replace the converse function

    
def eliza_chat():
    print("Listener\n---------")
    print("Talk to the program by typing in plain English, using normal upper-")
    print('and lower-case letters and punctuation.  Enter "quit" when done.')
    print("=" * 72)
    print("Hello. How was your class today?")

    eliza_chatbot.converse()
    

def demo():
    eliza_chat()
    

if __name__ == "__main__":
    demo()
```

```python
#CSC620
#HA#4
#Step 1 -- Bigram LM
#Paula Abigail Tam <921850992>

import nltk
from nltk.corpus import treebank
from nltk.util import bigrams
from collections import Counter

def corpus_loader():
    text = treebank.words('wsj_0088.mrg') #corpus used in HA#3
    normalized_text = normalize_words(text)
    return normalized_text #is a list

def normalize_words(corpus_tokens):
    new_tokens = [word.casefold() for word in corpus_tokens if word.isalnum()] #only alphanumeric words
    return new_tokens

def bigram_maker(tokens):
    bigrams_tokens = bigrams(tokens)
    return bigrams_tokens #returns bigrams

def count_words(tokens):
    corpus_word_count = Counter(tokens) #to count the individual words
    return corpus_word_count

new_dict = {} #dictionary to store the bigrams and probabilities

#all of this is for debugging? output checking
#nevermind, used it to get all the result
def all_counts():
    bigram_counts = Counter(bigram_maker(corpus_loader())) #dictionary of all bigrams and their counts
    w1_counts = list(count_words(corpus_loader()).items())
    w1_in_bigram = list(bigram_counts.items())
    for x in w1_in_bigram: #iterate bigrams
        for y in w1_counts: #iterate word counts
            if x[0][0].casefold() == y[0].casefold():
                new_dict[x[0]] = (x[1]/y[1]) #value is the (bigram count / first word count)
                #print(x[1], y[1], new_dict[x[0]]) #to test out the probabilities
    return new_dict
    #print(list(bigram_counts.items())) #can get individual keys, first element
    #print("bigrams: ", bigram_counts) #checking bigrams
    #print("words: ", w1_counts) #checking words

if __name__ == '__main__':
    all_counts()
    
```

```python
#CSC620
#HA#4
#Step 2 -- Word Predictor
#Paula Abigail Tam <921850992>

#I had to do
#pip install ipynb
#in order to get the ipynb import to work

import ipynb.fs.full.HA4_1_Paula_Tam
import re
from ipynb.fs.full.HA4_1_Paula_Tam import all_counts, corpus_loader
#all_counts returns a dictionary with all the bigrams and their probabilities
#corpus_loader returns the normalized text in the corpus

def filter_bigrams(user_input):
    regex = '{}$'.format(re.escape(user_input)) #using regex so it just matches with the whole input
    #e.g. if my user_input is "the", it'll match with "the", "then", "another", etc.
    #so I used the solution to match with ONLY the user_input
    #https://stackoverflow.com/questions/45696522/python-find-entire-word-in-string-using-regex-and-user-input
    
    word_filter = lambda w1: re.match(regex, w1[0][0], re.IGNORECASE) #insensitive matching just in case
    #w1[0] to match with the dictionary key --> w1[0][0] for the first element of the key (which is a tuple)
    
    filtered_dict = dict(filter(word_filter, all_counts().items())) #filters the dict of bigrams to only have the specified word_filter (aka the user_input)
    return filtered_dict #returns filtered dictionary

def sort_filtered_bigrams(filtered_dict):
    sorted_bigrams = sorted(filtered_dict.items(),key=lambda w: w[1], reverse=True) #double checking that it filters & sorts descending as a list
    return sorted_bigrams
    #print(sorted_bigrams[0][0][1]) #[0] first item, [0][0] -> key (tuple), [0][0][1] -> second item in key tuple
    
def print_out_results(sorted_bigrams):
    print("Possible next words:")
    for w in sorted_bigrams:
        print(w[0][1], ": ", round(w[1], 5)) #rounded to 5 decimal places just for better readability in output
    
def take_user_input():
    user_input = input("Enter a word to find bigrams of: ")
    if user_input in corpus_loader():
        filter_input = filter_bigrams(user_input)
        sorted_input = sort_filtered_bigrams(filter_input)
        print_out_results(sorted_input)
    else:
        print("This corpus does not contain this word.")

if __name__ == '__main__':
    take_user_input()
    #print(all_counts()) #double checking that the import worked
```
